#scoring

```python
for index, row in data.iterrows():
    if 1053.077 <= row['구독자'] < 12259.62:
        data.at[index, '구독자점수'] = 0
    elif 12259.62 <= row['구독자'] < 37407.69:
        data.at[index, '구독자점수'] = 3.75
    elif 37407.69 <= row['구독자'] < 88925:
        data.at[index, '구독자점수'] = 7.5
    elif row['구독자'] >= 88925:
        data.at[index, '구독자점수'] = 10
    else:
       pass
```

```python
data
```

```python
for index, row in data.iterrows():
    if  row['평균 조회수'] < 2712.698214:
        data.at[index, '조회수점수'] = 3
    elif 2712.698214 <= row['평균 조회수'] < 7245.525641:
        data.at[index, '조회수점수'] = 6
    elif 7245.525641 <= row['평균 조회수'] < 15566.12404:
        data.at[index, '조회수점수'] = 9
    elif 15566.12404 <= row['평균 조회수']< 33979.7326:
        data.at[index, '조회수점수'] = 12
    elif 33979.7326 <= row['평균 조회수']:
       data.at[index, '조회수점수'] = 15
    else:
      data.at[index, '조회수점수'] = 0
```

```python
data
```

```python
for index, row in data.iterrows():
    if  row['구독자'] < 12259.62:
        data.at[index, '구독자점수'] = 0
    elif 12259.62 <= row['구독자'] < 37407.69:
        data.at[index, '구독자점수'] = 3.75
    elif 37407.69 <= row['구독자'] < 88925:
        data.at[index, '구독자점수'] = 7.5
    elif 88925 <= row['구독자'] < 211615.4:
        data.at[index, '구독자점수'] = 11.25
    else:
        data.at[index, '구독자점수'] = 15
```

```python
data
```

```python
for index, row in data.iterrows():
    if  row['충성시청자_비율'] < 0.047295:
        data.at[index, '충성도점수'] = 2
    elif 0.047295 <= row['충성시청자_비율'] < 0.101182:
        data.at[index, '충성도점수'] = 4
    elif 0.101182 <= row['충성시청자_비율'] < 0.155089:
        data.at[index, '충성도점수'] = 6
    elif 0.155089 <= row['충성시청자_비율']< 0.305601:
        data.at[index, '충성도점수'] = 8
    elif 0.305601<= row['충성시청자_비율']:
       data.at[index, '충성도점수'] = 10
    else:
      data.at[index, '충성도점수'] = 0
```
