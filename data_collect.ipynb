# 댓글 데이터 통합 파일 생성

```python
total_comment_df = pd.DataFrame(columns=['youtube_id', 'comment', 'like_num', 'comment_date', 'owner_channel', 'video_num'])
```

```python
import os
import pandas as pd
for name in os.listdir('./data/댓글_데이터'):
    for video in os.listdir(f'./data/댓글_데이터/{name}'):
        temp = pd.read_csv(f'./data/댓글_데이터/{name}/{video}', index_col=0)
        owner = name
        video_num = video[len(name)+1:-17]
        temp['owner_channel'] = owner
        temp['video_num'] = video_num
        total_comment_df = pd.concat([total_comment_df, temp], axis=0)
        print(video + '병합 완료')
```

```python
total_comment_df.reset_index(drop=True, inplace=True)
total_comment_df.to_csv('./data/total_comment_data.csv', encoding='utf-8-sig')
```

```python
total_comment_df.to_excel('./data/total_comment_data.xlsx')
```

```python
total_comment_df.info()
```

```python
total_comment_df.loc[total_comment_df['video_num'].str.len() > 3, 'video_num'] = total_comment_df.loc[total_comment_df['video_num'].str.len() > 3, 'video_num'].str.split('_').str[-1]
```

```python
total_comment_df['video_num'] = total_comment_df['video_num'].astype(int)
```

```python
total_comment_df.info()
```

```python
total_comment_df.reset_index(drop=True, inplace=True)
total_comment_df.to_csv('./data/total_comment_data.csv', encoding='utf-8-sig')
```

```python
total_comment_df.to_excel('./data/total_comment_data.xlsx')
```

# 전체 영상정보 통합 파일 생성

```python
total_video_df = pd.DataFrame(columns=['제목', 'URL', '조회수', '업데이트 날짜', '영상 길이', '영상 좋아요 수', '채널명'])
```

```python
import os
for video in os.listdir('./data/total_video_data'):
        temp = pd.read_csv(f'./data/total_video_data/{video}', index_col=0)
        owner = video[:-15]
        temp['채널명'] = owner
        total_video_df = pd.concat([total_video_df, temp], axis=0)
        print(video + '병합 완료', os.listdir('./data/total_video_data').index(video))
```

# 1년치 영상정보 통합 파일 생성

```python
total_year_video_df = pd.DataFrame(columns=['제목', 'URL', '조회수', '업데이트 날짜', '영상 길이', '영상 좋아요 수', '채널명'])
```

```python
import os
for video in os.listdir('./data/1년치_영상정보'):
        temp = pd.read_csv(f'./data/1년치_영상정보/{video}', index_col=0)
        temp = temp[(temp['업데이트 날짜'] > '2022-08-01') & (temp['업데이트 날짜'] < '2023-09-01')]
        owner = video[:-15]
        temp['채널명'] = owner
        total_year_video_df = pd.concat([total_year_video_df, temp], axis=0)
        print(video + '병합 완료', os.listdir('./data/1년치_영상정보').index(video))
```

```python
total_year_video_df.info()
```

```python
total_year_video_df.reset_index(drop=True, inplace=True)
total_year_video_df.to_csv('./data/total_year_video_df.csv', encoding='utf-8-sig')
```

# 수집된 댓글 데이터 파일 개수와 1년치 영상정보의 행 개수가 일치하는지 확인

```python
not_enough_youtuber = []
for youtuber in os.listdir('./data/댓글_데이터'):
    com_file_num = len(os.listdir(f'./data/댓글_데이터/{youtuber}'))
    temp_df = pd.read_csv(f'./data/1년치_영상정보/{youtuber}_video_data.csv')
    temp_df = temp_df[(temp_df['업데이트 날짜'] > '2022-08-01') & (temp_df['업데이트 날짜'] < '2023-09-01')]
    if com_file_num != len(temp_df):
        print(youtuber, com_file_num, len(temp_df))
        not_enough_youtuber.append(youtuber)
```

```python
not_enough_youtuber = pd.DataFrame(not_enough_youtuber)
not_enough_youtuber.to_csv('./data/not_enough_youtuber.csv', index=False)
not_enough_youtuber
```
