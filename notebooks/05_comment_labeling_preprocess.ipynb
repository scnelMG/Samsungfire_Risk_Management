```python
import os
import pandas as pd
```

```python
os.listdir(f'./data/labeled_data')
```

```python
label_comment_df = pd.DataFrame(columns=['comment', 'label'])
```

```python
for file in os.listdir(f'./data/labeled_data')[5:16]:
    temp = pd.read_csv(f'./data/labeled_data/{file}', encoding='cp949')
    try:
        temp = temp[['comment', 'labeling']]
    except:
        temp = temp[['comment', 'labe;ing']]
    temp.columns = ['comment', 'label']
    label_comment_df = pd.concat([label_comment_df, temp], ignore_index=True)
label_comment_df
```

```python
# label_comment_df에서 label이 0인건 지우고, -1인건 0으로 변환
label_comment_df = label_comment_df[label_comment_df['label'] != 0]
label_comment_df['label'] = label_comment_df['label'].apply(lambda x: 0 if x == -1 else x)
```

```python
temp = pd.read_csv(f'./data/labeled_data/labeling3.csv', encoding='cp949')
temp = temp[['comment', 'labeling']]
temp.columns = ['comment', 'label']
temp['label'] = temp['label'].apply(lambda x: 0 if x == -1 else x)
label_comment_df = pd.concat([label_comment_df, temp], ignore_index=True)
label_comment_df
```

```python
label_comment_df.value_counts('label')
```

```python
temp = pd.read_excel(f'./data/labeled_data/labeling2.xlsx')
temp = temp[['comment', 'labeling']]
temp.columns = ['comment', 'label']
temp['label'] = temp['label'].apply(lambda x: 0 if x == -1 else x)
label_comment_df = pd.concat([label_comment_df, temp], ignore_index=True)
label_comment_df
```

```python
label_comment_df.value_counts('label')
```

```python
for file in os.listdir(f'./data/labeled_data')[2:5] + os.listdir(f'./data/labeled_data')[-2:]:
    temp = pd.read_excel(f'./data/labeled_data/{file}')
    temp = temp[['comment', '긍부정']]
    temp['긍부정'] = temp['긍부정'].apply(lambda x: 0 if x == -1 else x)
    temp.columns = ['comment', 'label']
    label_comment_df = pd.concat([label_comment_df, temp], ignore_index=True)
    label_comment_df
```

```python
label_comment_df
```

```python
# label이 결측치인 행 제거
label_comment_df = label_comment_df.dropna(axis=0)
label_comment_df
```

```python
# 중복 확인
label_comment_df.drop_duplicates(inplace=True)

# 인덱스 초기화
label_comment_df.reset_index(drop=True, inplace=True)
# 데이터 타입 변경
label_comment_df['label'] = label_comment_df['label'].astype(int)
label_comment_df
```

```python
label_comment_df['label'].value_counts()
```

```python
label_comment_df.to_csv('./data/labeled_data/label_comment_df.csv', index=False)
```
