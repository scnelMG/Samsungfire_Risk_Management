```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('final_all_data (1).xlsx')
```

```python
data.head(10)
```

```python
unique_youtubers = data['유투버'].unique()

plt.figure(figsize=(10, 6))

for youtuber in unique_youtubers:
    youtuber_data = data[data['유투버'] == youtuber]
    x = youtuber_data['연도'].astype(str) + '-' + youtuber_data['월'].astype(str)
    y = youtuber_data['최종_점수']
    
    plt.plot(x, y, label=youtuber, color='black') 
    plt.xlabel('연도-월')
    plt.ylabel('최종_점수')
    plt.ylim(0, 100)  # y축 범위를 0에서 100으로 고정
    plt.title('유투버별 총 점수 시계열 그래프')
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid(True)
    plt.show()
```
