```python
# 날짜 컬럼을 새로운 열로 추가
fol_melted = pd.melt(followers, id_vars=['유투버_명'], var_name='date', value_name='구독자')

# 결과 출력
fol_melted
```

```python
fol_melted['date'] = pd.to_datetime(fol_melted['date'])
```

```python
fol_melted['연도'] = fol_melted['date'].dt.year
fol_melted['월'] = fol_melted['date'].dt.month

fol_melted = fol_melted.rename(columns={'유투버_명': '채널명'})
```

```python
fol_melted
```
