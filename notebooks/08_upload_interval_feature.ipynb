```python
start_date = [pd.to_datetime('2021-08-01'), pd.to_datetime('2021-09-01'), pd.to_datetime('2021-10-01'), pd.to_datetime('2021-11-01'), pd.to_datetime('2021-12-01'), pd.to_datetime('2022-01-01'), pd.to_datetime('2022-02-01'), pd.to_datetime('2022-03-01'),
                pd.to_datetime('2022-04-01'), pd.to_datetime('2022-05-01'), pd.to_datetime('2022-06-01'), pd.to_datetime('2022-07-01'), pd.to_datetime('2022-08-01'),
                pd.to_datetime('2022-09-01'), pd.to_datetime('2022-10-01'), pd.to_datetime('2022-11-01'), pd.to_datetime('2022-12-01'), pd.to_datetime('2023-01-01'), pd.to_datetime('2023-02-01'), pd.to_datetime('2023-03-01'),
                pd.to_datetime('2023-04-01'), pd.to_datetime('2023-05-01'), pd.to_datetime('2023-06-01'), pd.to_datetime('2023-07-01'), pd.to_datetime('2023-08-01')]

end_date = [pd.to_datetime('2021-08-31'), pd.to_datetime('2021-09-30'), pd.to_datetime('2021-10-31'), pd.to_datetime('2021-11-30'), pd.to_datetime('2021-12-31'), pd.to_datetime('2022-01-31'), pd.to_datetime('2022-02-28'), pd.to_datetime('2022-03-31'),
                pd.to_datetime('2022-04-30'), pd.to_datetime('2022-05-31'), pd.to_datetime('2022-06-30'), pd.to_datetime('2022-07-31'), pd.to_datetime('2022-08-31'),
             pd.to_datetime('2022-09-30'), pd.to_datetime('2022-10-31'), pd.to_datetime('2022-11-30'), pd.to_datetime('2022-12-31'), pd.to_datetime('2023-01-31'), pd.to_datetime('2023-02-28'), pd.to_datetime('2023-03-31'),
                pd.to_datetime('2023-04-30'), pd.to_datetime('2023-05-31'), pd.to_datetime('2023-06-30'), pd.to_datetime('2023-07-31'), pd.to_datetime('2023-08-31')]

for name in df_frame['유투버'].unique():
    print(name)
    date_df = total_video[total_video['채널명'] == name][['업데이트 날짜','채널명']].sort_values(by='업데이트 날짜', ascending=True)
    temp = []
    temp_num = []
    for i in range(len(start_date)):
        filtered_df = date_df[(date_df['업데이트 날짜'] >= start_date[i]) & (date_df['업데이트 날짜'] <= end_date[i])]
        # 업데이트 날짜의 간격 일자를 구하기
        filtered_df['term'] = filtered_df['업데이트 날짜'].diff()
        filtered_df['term'] = filtered_df['term'].dt.days
        mean_diff = filtered_df['term'].mean()
        temp.append(mean_diff)
        temp_num.append(len(filtered_df))
    df_frame.loc[df_frame['유투버'] == name, '평균_영상간격'] = temp
    df_frame.loc[df_frame['유투버'] == name, '영상_개수'] = temp_num
```

```python
term_df = pd.read_excel('./data/간격_민규.xlsx')
term_df['null_지속'] = np.nan
term_df
```

```python
term_df['null_지속_temp'] = 0
```

```python
# 평균_영상간격이 nan이고 유투버가 @yekooooong인 행 추출
for name in term_df['유투버'].unique():
    print(name)
    a = term_df[term_df['유투버'] == name]
    a['null_지속_temp'] = 0
    a.loc[(a['평균_영상간격'].isnull()) & (a['유투버'] == name), 'null_지속_temp'] = 1
    a['null_지속'] = a['null_지속_temp'].groupby((a['null_지속_temp'] != a['null_지속_temp'].shift()).cumsum()).cumcount() + 1
    a.loc[a['null_지속_temp'] == 0, 'null_지속'] = 0
    term_df.loc[term_df['유투버'] == name, 'null_지속'] = a['null_지속']
term_df
```
